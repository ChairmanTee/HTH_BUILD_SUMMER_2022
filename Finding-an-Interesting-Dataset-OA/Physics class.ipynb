```python
import matplotlib.pyplot as plt
import math
import numpy as np

x = x1 = []
y = y1 = []
Vyy = []
Vxx = []
ti = []
Θ = 45
V = 20
g = 9.8
t = 0.0
Δt = 0.001
a = 0.0
b = 0.0
n = 1000
Vx = V*np.cos(Θ*math.pi/100)
Vy = V*np.sin(Θ*math.pi/100)
t = (2 * V * np.sin(Θ*math.pi/100)) / g
Vyy.append(Vy)
Vxx.append(Vx)
ti.append(t)

for i in range(0,n):
    # Vxx = Vxx
    t = t + i * Δt
    ti.append(t)
    a = a + Vxx * Δt
    x.append(a)
    Vyy= Vyy - g * Δt
    Vyy.append(Vy)
    Vxx.append(Vx)
    b=b + Vyy * Δt
    y.append(b)

# print (x," ",y)

# plt.plot(x,y)
# plt.ylabel('Y (meters)')
# plt.xlabel('X (meters)')
# # # plt.axis([0,50,0,20])
# plt.show()


plt.plot(x,y)
plt.ylabel('Y (m)')
plt.xlabel('X (m)')
plt.title('Projectile Motion')
plt.axis([0,5,0,5])
plt.figure()
plt.plot(ti,Vyy, ti,Vxx)
plt.ylabel('Vx (m/s)')
plt.xlabel('T (s)')
plt.title('Velocity Vs Time')
plt.axis([0,600, 0,20])
plt.show()
```
