```python
import pandas
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
#pandas.options.display.max_columns = None
#pandas.options.display.max_rows = None

Population_data = pandas.read_csv("world_population_by_country.csv")

Population_data = Population_data[Population_data['Fert. Rate'].str.contains('N\.A\.') == False]
Population_data.dropna(subset=['Fert. Rate'],inplace = True)
# Population_data['Land Area (Km²)']= Population_data['Land Area (Km²)'].str.replace(",","").astype(int)
Population_data['Population 2020']= Population_data['Population 2020'].str.replace(",","").astype(int)
Population_data['World Share']= Population_data['World Share'].str.replace("%","").astype(float)
Population_data = Population_data[Population_data['Urban Pop %'].str.contains('N\.A\.') == False]
Population_data['Urban Pop %']= Population_data['Urban Pop %'].str.replace("%","").astype(int)
Population_data['Fert. Rate']= Population_data['Fert. Rate'].astype(float)
sns.set(font_scale = 1)

Population_data
```

```python
#What are the top 10 most populous countries in the world?

Population_data.head(10)
```

```python
# Piechart showing the top 10 most populous countries in the world
import matplotlib.pyplot as plt

plt.pie(x = Population_data['Population 2020'].head(10), labels = Population_data['Country (or dependency)'].head(10)) 
plt.show()
```

```python
# Barplot of the 10 most populous countries with their percentage of Urban Population

ax = sns.barplot( data = Population_data.head(10), x='Country (or dependency)', y='Population 2020', hue= 'Urban Pop %', dodge= False)
ax.set(xlabel='Country', ylabel='Population (Billions)', title='Population data',)

plt.setp(ax.get_xticklabels(), rotation=45)
ax.figure.set_size_inches(20,10)
sns.set(font_scale = 1)
sns.set_theme(style="darkgrid")
tips = sns.load_dataset("tips")
plt.ylim(0,2000000000)
plt.show()
```

```python
# Top 10 Countries with highest percentage of Urban Population
# Population_data = Population_data[Population_data['Urban Pop %'].str.contains('N\.A\.') == False]

sorted_population_data = Population_data.sort_values(by='Urban Pop %',ascending=False)

sorted_population_data.head(10)

ax = sns.barplot(data=sorted_population_data.head(10), x="Country (or dependency)", y="Urban Pop %", ci = None)

# sns.countplot(data=Population_data.head(10), x="Country (or dependency)")

plt.title('Countries with highest percentage of Urban Populaton')
plt.xlabel('Country')
plt.ylabel('Urban Pop (%)')
plt.xticks(rotation=45)
ax.figure.set_size_inches(20,10)
sns.despine()
plt.show()
```

```python
# TOP 10 Countries with lesst percentage of Urban Population
sorted_population_data = Population_data.sort_values(by='Urban Pop %',ascending=False)

sorted_population_data.tail(10)

ax = sns.barplot(data=sorted_population_data.tail(10), x="Country (or dependency)", y="Urban Pop %", ci = None)

# sns.countplot(data=Population_data.head(10), x="Country (or dependency)")

plt.title('Countries with least percentage of Urban Populaton')
plt.xlabel('Country')
plt.ylabel('Urban Pop (%)')
plt.xticks(rotation=45)
ax.figure.set_size_inches(20,10)
sns.despine()
plt.show()
```

```python
# What top 5 countries have the most land area and what top 5 countries have the least land area?
Population_data['Land Area (Km²)']= Population_data['Land Area (Km²)'].str.replace(",","").astype(int)

sorted_population_data = Population_data.sort_values(by='Land Area (Km²)',ascending=False)

sorted_population_data
```

```python
# top 5 countries with highest number of migrants and least number of migrants

Population_data.dropna(subset=['Migrants (net)'],inplace = True)

Population_data['Migrants (net)']= Population_data['Migrants (net)'].str.replace(",","").astype(int)

sorted_population_data = Population_data.sort_values(by='Migrants (net)',ascending=False)

sorted_population_data
```

```python
# Correllation between fertility rate and world share
Population_data['World Share']= Population_data['World Share'].str.replace("%","").astype(float)
Population_data['Fert. Rate']= Population_data['Fert. Rate'].astype(float)

sns.lineplot(x='World Share', y='Fert. Rate', data= Population_data)
```

```python
# Heatmap 
fig, ax = plt.subplots(figsize=(3,3))

sns.heatmap(Population_data[['Fert. Rate', 'World Share']].corr(), cmap="coolwarm", annot = True, square = True)
```

```python
# Correllation between fertility rate and world share
ax = sns.barplot(data = Population_data.head(10), x='World Share', y='Fert. Rate', hue = 'Country (or dependency)', dodge = False)
ax.set(xlabel='World Share (%)', ylabel='Fert. Rate', title='World Share vs Fert. Rate')
sns.set_theme(style="darkgrid")
ax.figure.set_size_inches(20,10)
plt.xticks(rotation=45)
plt.show()
```
